```python
# Chart 1: Top Selling Products
data = {
    'Product': ['Sunglasses', 'Yoga Mat', 'Wireless Mouse'],
    'Units Sold': [2, 2, 1]
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Units Sold', data=df, palette='viridis')
plt.title('Top 3 Selling Products')
plt.ylabel('Units Sold')
plt.xlabel('Product')
plt.tight_layout()
plt.show()
```

```python
revenue_data = {
    'Region': ['Delhi', 'Mumbai', 'Bangalore'],
    'Revenue': [2398, 1998, 1299]
}

df_rev = pd.DataFrame(revenue_data)

plt.figure(figsize=(6, 6))
plt.pie(df_rev['Revenue'], labels=df_rev['Region'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Revenue Share by Region')
plt.axis('equal')
plt.show()
```

```python
funnel = {
    'Stage': ['Visited', 'Added to Cart', 'Purchased'],
    'Users': [4, 2, 3]
}

df_funnel = pd.DataFrame(funnel)

plt.figure(figsize=(7, 5))
sns.barplot(x='Stage', y='Users', data=df_funnel, palette='Greens_d')
plt.title('E-Commerce Funnel Stages')
plt.ylabel('Number of Users')
plt.xlabel('Funnel Stage')
plt.tight_layout()
plt.show()
```
